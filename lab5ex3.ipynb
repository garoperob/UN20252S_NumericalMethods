```python
import numpy as np
import pandas as pd
import sympy as sy
import matplotlib.pyplot as plt
```

```python
def simpson(a, b, mathFunction, m):
    h = (b-a)/(2*m)
    print(h)
    fa = mathFunction(a)
    fb = mathFunction(b)

    even = 0
    odd = 0
    x_intervals = [a]
    y_intervals = [mathFunction(a)]

    for k in range(1,2*m):
        xVal = a+k*h
        x_intervals.append(xVal)
        yVal = mathFunction(xVal)
        y_intervals.append(yVal)
        if k % 2 == 0:
            even += yVal
        else:
            odd += yVal
    x_intervals.append(b)
    y_intervals.append(mathFunction(b))

    result = (h/3) * (fa + fb) + (2*(h/3)) * even + (4*(h/3)) * odd

    return [result, x_intervals, y_intervals]


    
```

```python
expr = input("Enter the function: ")
var = input("Enter the variable: ")
a = float(input("Enter the lower bound: "))
b = float(input("Enter the upper bound: "))
m = int(input("Enter the total nodes you want to estimate the result: "))
h = eval(f"lambda {var}:{expr}")

integral = simpson(a,b,h,m)

xCuts = integral[1]
yResult = integral[2]
xPoints = np.linspace(a,b,100)
yPoints = [h(x) for x in xPoints]
plt.plot(xPoints, yPoints)
plt.fill_between(xCuts, yResult, color= 'orange', alpha = 0.3)
plt.grid(True)

print(expr)
print(integral[0])
```
